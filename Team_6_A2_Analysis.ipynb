```python
# Pre-defined functions 
########################################
# scree_plot
########################################
def scree_plot(pca_object, export = False):
    """
    Visualizes a scree plot from a pca object.
    
    PARAMETERS
    ----------
    pca_object | A fitted pca object
    export     | Set to True if you would like to save the scree plot to the
               | current working directory (default: False)
    """
    # building a scree plot

    # setting plot size
    fig, ax = plt.subplots(figsize=(10, 8))
    features = range(pca_object.n_components_)


    # developing a scree plot
    plt.plot(features,
             pca_object.explained_variance_ratio_,
             linewidth = 2,
             marker = 'o',
             markersize = 10,
             markeredgecolor = 'black',
             markerfacecolor = 'grey')


    # setting more plot options
    plt.title('Scree Plot')
    plt.xlabel('PCA feature')
    plt.ylabel('Explained Variance')
    plt.xticks(features)

    if export == True:
    
        # exporting the plot
        plt.savefig('./__analysis_images/top_customers_correlation_scree_plot.png')
        
    # displaying the plot
    plt.show()


########################################
# unsupervised_scaler
########################################
def unsupervised_scaler(df):
    """
    Standardizes a dataset (mean = 0, variance = 1). Returns a new DataFrame.
    Requires sklearn.preprocessing.StandardScaler()
    
    PARAMETERS
    ----------
    df     | DataFrame to be used for scaling
    """

    # INSTANTIATING a StandardScaler() object
    scaler = StandardScaler()


    # FITTING the scaler with the data
    scaler.fit(df)


    # TRANSFORMING our data after fit
    x_scaled = scaler.transform(df)

    
    # converting scaled data into a DataFrame
    new_df = pd.DataFrame(x_scaled)


    # reattaching column names
    new_df.columns = df.columns
    
    return new_df
```

In our next step, we shall merge all the scaled dataframes into a new dataframe called psychometric. As the individual psychometric answers have been scaled respondent-wise, we may not scale it again. Then, we fit PCA on the aggregated dataframe.

```python
# Concating the three dataframes
psychometric_df = pd.concat([tech_scaled, personality_scaled,spending_scaled], axis = 1)
psychometric_df.head()
```

```python
# INSTANTIATING a PCA object with no limit to principal components
pca = PCA(n_components = None,
          random_state = 219)


# FITTING and TRANSFORMING the psychometric df
psychometric_pca = pca.fit_transform(psychometric_df)


# calling the scree_plot function
scree_plot(pca_object = pca)
```

We shall reduce the number of principal components to a reasonable number based on the scree plot above. We assume five PCs is a reasonable number based on the elbow in the scree plot.  

```python
# component number counter
component_number = 0

# looping over each principal component
for variance in pca.explained_variance_ratio_:
    component_number += 1
    
    print(f"PC {component_number}: {variance.round(3)}")
```

```python
# printing the sum of all explained variance ratios
print(pca.explained_variance_ratio_.sum())
```

It would have been reasonable to retain enough PCs so that the cumulative explained variance ratio is greater than or equal to 0.80. However, we shall proceed with 5 PCs as the number of PCs is 40. The 5 PCs give a cumulative explained variance ratio of around 0.30.

```python
# INSTANTIATING a new model using the first five principal components
pca_5 = PCA(n_components = 5,
            random_state = 219)


# FITTING and TRANSFORMING the purchases_scaled
psychometric_pca_5 = pca_5.fit_transform(psychometric_df)


# calling the scree_plot function
scree_plot(pca_object = pca_5)
```

We run the following code to compare the variance of the unlimited PCA model with the variance of the reduced PCA model. We are doing this in this script simply to show that the explained variance for each principal component does not change after dropping smaller PCs.

```python
####################
### Max PC Model ###
####################
# transposing pca components (pc = MAX)
factor_loadings = pd.DataFrame(np.transpose(pca.components_))


# naming rows as original features
factor_loadings = factor_loadings.set_index(psychometric_df.columns)


##################
### 5 PC Model ###
##################
# transposing pca components (pc = 5)
factor_loadings_5 = pd.DataFrame(np.transpose(pca_5.components_))


# naming rows as original features
factor_loadings_5 = factor_loadings_5.set_index(psychometric_df.columns)


# checking the results
print(f"""
MAX Components Factor Loadings
------------------------------
{factor_loadings.round(2)}


5 Components Factor Loadings
------------------------------
{factor_loadings_5.round(2)}
""")
```

<br><h2><strong>Part IV - Clustering</strong></h2><br>
After having run the analysis separately for both databases already separated in psychometrics and app-affinity, it is time to join them and then begin to form our clusters. These are basically groups that have the same values and are classified by the algorithm in the same group. This way of classifying makes the identification of these groupings much easier to then start making decisions.

```python
# checking variance amongst personas
np.var(app_users_df)
```

```python
# applying the unsupervised_scaler function
pca_scaled_app_users = unsupervised_scaler(df = app_users_df)


# checking pre- and post-scaling variance
print(np.var(app_users_df), '\n\n')
print(np.var(pca_scaled_app_users))
```

```python
# grouping data based on Ward distance
standard_mergings_ward = linkage(y = pca_scaled_app_users,
                                 method = 'ward',
                                 optimal_ordering = True)


# setting plot size
fig, ax = plt.subplots(figsize=(12, 12))

# developing a dendrogram
dendrogram(Z = standard_mergings_ward,
           leaf_rotation = 90,
           leaf_font_size = 6)


# rendering the plot
plt.show()
```

```python
# INSTANTIATING a k-Means object with 5 clusters
app_users_k_pca = KMeans(n_clusters   = 5,
                         random_state = 219)


# fitting the object to the data
app_users_k_pca.fit(pca_scaled_app_users)


# converting the clusters to a DataFrame
app_users_kmeans_pca = pd.DataFrame({'Cluster': app_users_k_pca.labels_})


# checking the results
print(app_users_kmeans_pca.iloc[: , 0].value_counts())
```

```python
# storing cluster centers
centroids_pca = app_users_k_pca.cluster_centers_


# converting cluster centers into a DataFrame
centroids_pca_df = pd.DataFrame(centroids_pca)


# renaming principal components
centroids_pca_df.columns = ['young_starter',   
                           'parent',       
                           'non-tech spender',
                            'shopper',
                            'thrifty-techie','simpleton',   
                           'movie_buff',       
                           'freeby_lover',
                            'netflix_netizen',
                            'social_networker' ]


# checking results (clusters = rows, pc = columns)
centroids_pca_df.round(2)
```

```python
# concatenating cluster memberships with principal components
clst_pca_df = pd.concat([app_users_kmeans_pca,
                         app_users_df],
                         axis = 1)


# concatenating demographic information with pca-clusters
final_pca_clust_df = pd.concat([mob_app.loc[ : , ['Age', 'edn_lvl', 'mrtl_stts', 
                                                  'no_kids','kids_<6','kids_6-12', 
                                                  'kids_13-17', 'kids_>18', 
                                                  'race', 'if_Latino', 'income',
                                                  'if_male']],
                                             clst_pca_df.round(decimals = 2)],
                                              axis = 1)


# renaming columns
final_pca_clust_df.columns = ['Age', 'edn_lvl', 'mrtl_stts', 'no_kids','kids_<6',
                              'kids_6-12', 'kids_13-17', 'kids_>18', 'race', 
                              'if_Latino', 'income', 'if_male', 'Cluster',
                              'young_starter', 'parent', 'non-tech spender', 
                              'shopper', 'thrifty-techie', 'simpleton', 'movie_buff', 
                              'freeby_lover', 'netflix_netizen', 'social_networker' ]


# checking the results
print(final_pca_clust_df.head(n = 5))
print(final_pca_clust_df.info())
```
